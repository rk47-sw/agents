```python
# Start with imports - ask ChatGPT to explain any package that you don't know

import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from anthropic import Anthropic
from IPython.display import Markdown, display
```

```python
# Always remember to do this!
load_dotenv(override=True)
```

```python
# Print the key prefixes to help with any debugging
# GEMINI_API_KEY is required - get one free at https://aistudio.google.com/
# All others are optional (only needed if you want that model as a competitor)

gemini_api_key = os.getenv('GEMINI_API_KEY')
openai_api_key = os.getenv('OPENAI_API_KEY')
anthropic_api_key = os.getenv('ANTHROPIC_API_KEY')
deepseek_api_key = os.getenv('DEEPSEEK_API_KEY')
groq_api_key = os.getenv('GROQ_API_KEY')

if gemini_api_key:
    print(f"Gemini API Key exists and begins {gemini_api_key[:8]}")
else:
    print("Gemini API Key not set - this is REQUIRED!")

if openai_api_key:
    print(f"OpenAI API Key exists and begins {openai_api_key[:8]}")
else:
    print("OpenAI API Key not set (optional - needed for GPT competitor)")

if anthropic_api_key:
    print(f"Anthropic API Key exists and begins {anthropic_api_key[:7]}")
else:
    print("Anthropic API Key not set (optional - needed for Claude competitor)")

if deepseek_api_key:
    print(f"DeepSeek API Key exists and begins {deepseek_api_key[:3]}")
else:
    print("DeepSeek API Key not set (optional)")

if groq_api_key:
    print(f"Groq API Key exists and begins {groq_api_key[:4]}")
else:
    print("Groq API Key not set (optional)")
```

```python
request = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. "
request += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": request}]
```

```python
messages
```

```python
# Create the Gemini orchestrator using the OpenAI-compatible endpoint.
# This is our primary driver for question generation, judging, and reflection.

gemini_orchestrator = OpenAI(
    api_key=gemini_api_key,
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

response = gemini_orchestrator.chat.completions.create(
    model="gemini-2.5-flash",
    messages=messages,
)
question = response.choices[0].message.content
print(question)
```

```python
competitors = []
answers = []
messages = [{"role": "user", "content": question}]
```

## Competitors

Below we send the same question to multiple LLMs. Each competitor is optional — skip any cell if you don't have that API key. Gemini 2.5 Flash (via the orchestrator) will judge all responses at the end.


```python
# Competitor 1: OpenAI GPT (optional - requires OPENAI_API_KEY)

model_name = "gpt-4.1-mini"

openai = OpenAI()
response = openai.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Competitor 2: Anthropic Claude (optional - requires ANTHROPIC_API_KEY)
# Anthropic has a slightly different API, and max_tokens is required

model_name = "claude-sonnet-4-5"

claude = Anthropic()
response = claude.messages.create(model=model_name, messages=messages, max_tokens=1000)
answer = response.content[0].text

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Competitor 3: Gemini 2.5 Flash - we reuse the orchestrator client since it's already configured

model_name = "gemini-2.5-flash"

response = gemini_orchestrator.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Competitor 4: DeepSeek (optional - requires DEEPSEEK_API_KEY)

deepseek = OpenAI(api_key=deepseek_api_key, base_url="https://api.deepseek.com/v1")
model_name = "deepseek-chat"

response = deepseek.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# Competitor 5: Groq (optional - requires GROQ_API_KEY)

groq = OpenAI(api_key=groq_api_key, base_url="https://api.groq.com/openai/v1")
model_name = "llama-3.3-70b-versatile"

response = groq.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

## Don't have Ollama? Use Gemini Flash Lite instead

If you don't want to run a local model, `gemini-2.0-flash-lite` is Google's lightweight, low-latency model — a great free cloud alternative to a local Ollama model. It uses the same `gemini_orchestrator` client, so no extra setup is needed.


```python
# Competitor 7 (Ollama alternative): Gemini 2.0 Flash Lite
# Fast, lightweight, and free - a great stand-in if you don't have Ollama running locally

model_name = "gemini-2.5-flash-lite"

response = gemini_orchestrator.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content

display(Markdown(answer))
competitors.append(model_name)
answers.append(answer)
```

```python
# So where are we?

print(competitors)
print(answers)
```

```python
# It's nice to know how to use "zip"
for competitor, answer in zip(competitors, answers):
    print(f"Competitor: {competitor}\n\n{answer}")
```

```python
# Let's bring this together - note the use of "enumerate"

together = ""
for index, answer in enumerate(answers):
    together += f"# Response from competitor {index+1}\n\n"
    together += answer + "\n\n"
```

```python
print(together)
```

```python
judge = f"""You are judging a competition between {len(competitors)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
print(judge)
```

```python
judge_messages = [{"role": "user", "content": judge}]
```

```python
# Judgement time! - Gemini 2.5 Flash acts as the impartial judge

response = gemini_orchestrator.chat.completions.create(
    model="gemini-2.5-flash",
    messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
```

```python
# OK let's turn this into results!

results_dict = json.loads(results)
ranks = results_dict["results"]
for index, result in enumerate(ranks):
    competitor = competitors[int(result)-1]
    print(f"Rank {index+1}: {competitor}")
```

## Exercise Answer

**Patterns used so far:**
1. **Parallelization** — the same question is sent to multiple LLM competitors simultaneously, and results are gathered together.
2. **LLM-as-Judge** — a separate LLM evaluates and ranks all the responses, acting as an impartial arbiter.

**Adding a third pattern: Reflection**  
We take the winning response and ask the winning model to *critique* its own answer, then produce an improved version. This is the **Reflection / Self-Critique** agentic pattern — the agent reviews its own output and iteratively refines it.


```python
# Reflection Pattern - Step 1: identify the winner and their original answer

winner_index = int(ranks[0]) - 1
winning_model = competitors[winner_index]
winning_answer = answers[winner_index]

print(f"Winner: {winning_model}")
print(f"\nOriginal answer:\n{winning_answer}")
```

```python
# Reflection Pattern - Step 2: ask Gemini to critique the winning answer

critique_prompt = f"""You were asked this question:

{question}

You gave this answer:

{winning_answer}

Please critically evaluate your own answer. Identify:
1. Any weaknesses, gaps, or logical flaws
2. Points that could be argued more precisely or with more nuance
3. Anything important that was omitted

Be honest and rigorous in your self-critique."""

critique_messages = [{"role": "user", "content": critique_prompt}]

response = gemini_orchestrator.chat.completions.create(
    model="gemini-2.5-flash",
    messages=critique_messages
)
critique = response.choices[0].message.content

display(Markdown("### Self-Critique\n\n" + critique))
```

```python
# Reflection Pattern - Step 3: Gemini produces a refined answer based on the critique

refine_prompt = f"""You were asked this question:

{question}

Your original answer was:

{winning_answer}

You then critiqued your answer and identified these issues:

{critique}

Now write an improved, refined version of your answer that addresses all the weaknesses you identified. 
Make it clearer, more complete, and more rigorous."""

refine_messages = [{"role": "user", "content": refine_prompt}]

response = gemini_orchestrator.chat.completions.create(
    model="gemini-2.5-flash",
    messages=refine_messages
)
refined_answer = response.choices[0].message.content

display(Markdown("### Refined Answer (after Reflection)\n\n" + refined_answer))
```
