```python
# Check the key - we're using Google Gemini. Get your free API key at https://aistudio.google.com/

import os
gemini_api_key = os.getenv('GEMINI_API_KEY')

if gemini_api_key:
    print(f"Gemini API Key exists and begins {gemini_api_key[:8]}")
else:
    print("Gemini API Key not set - please head to the troubleshooting guide in the setup folder")
```

```python
# And now - the all important import statement
# If you get an import error - make sure you ran: uv add google-genai
# Then restart the kernel and try again

from google import genai
```

```python
# And now we'll create a Gemini client using our API key
# If you're not sure what it means to create an instance of a class - head over to the guides folder (guide 6)!
# If you get a NameError - head over to the guides folder (guide 6) to learn about NameErrors - always instantly fixable

client = genai.Client(api_key=gemini_api_key)
```

```python
# Create a prompt - Gemini's generate_content API accepts plain strings or lists of strings

messages = ["What is 2+2?"]
```

```python
# And now call it! Any problems, head to the troubleshooting guide
# This uses gemini-2.0-flash, a fast and very capable model from Google
# You can also try gemini-2.5-pro for more advanced reasoning

response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=messages
)

print(response.text)
```

```python
# And now - let's ask for a challenging question:

question = "Please propose a hard, challenging question to assess someone's IQ. Respond only with the question."
```

```python
# Ask Gemini to generate a hard question - gemini-2.0-flash is fast and capable

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=question
)

question = response.text

print(question)
```

```python
# Now pass the generated question back to Gemini to answer it

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=question
)

answer = response.text
print(answer)
```

```python
from IPython.display import Markdown, display

display(Markdown(answer))

```

```python
# Step 1: Ask Gemini to suggest a business area ripe for Agentic AI

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents="Pick one specific business area that would be worth exploring for an Agentic AI opportunity. Respond with just the business area name and a one-sentence explanation."
)

business_area = response.text
print("Business Area:\n", business_area)
```

```python
# Step 2: Ask Gemini to identify a specific pain-point in that business area

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=f"Given this business area: {business_area}\n\nDescribe one specific, challenging pain-point in that industry that is currently difficult to solve and would be ripe for an Agentic AI solution. Be concrete and specific."
)

pain_point = response.text
print("Pain Point:\n", pain_point)
```

```python
# Step 3: Ask Gemini to propose an Agentic AI solution for that pain-point

response = client.models.generate_content(
    model="gemini-2.5-flash-lite",
    contents=f"Business area: {business_area}\n\nPain-point: {pain_point}\n\nPropose a concrete Agentic AI solution for this pain-point. Describe what autonomous agents would do, how they would work together, and what the business impact would be."
)

solution = response.text
print("Agentic AI Solution:\n", solution)
```

```python
# Display the full result nicely formatted

from IPython.display import Markdown, display

display(Markdown(f"## Business Area\n{business_area}\n\n## Pain Point\n{pain_point}\n\n## Proposed Agentic AI Solution\n{solution}"))
```
