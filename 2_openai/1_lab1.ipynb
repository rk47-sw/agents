```python
# The imports

import os
from dotenv import load_dotenv
from openai import AsyncOpenAI
from agents import Agent, Runner, trace, set_default_openai_client, set_default_openai_api
```

```python
load_dotenv(override=True)

# Point the Agents SDK at Gemini's OpenAI-compatible endpoint
# set_default_openai_api("chat_completions") is required because Gemini doesn't
# support the Responses API that the SDK uses by default

gemini_client = AsyncOpenAI(
    api_key=os.getenv("GEMINI_API_KEY"),
    base_url="https://generativelanguage.googleapis.com/v1beta/openai/"
)

set_default_openai_client(gemini_client)
set_default_openai_api("chat_completions")
```

```python
# Make an agent with name, instructions, model
# Everything else stays exactly the same - the SDK is model-agnostic!

agent = Agent(name="Jokester", instructions="You are a joke teller", model="gemini-2.5-flash")
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

with trace("Telling a joke"):
    result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
    print(result.final_output)
```

## Note on tracing

Tracing with `trace()` still works as a context manager (it's a no-op when not connected to OpenAI's platform).  
If you want to see traces, you can use OpenAI's trace platform at https://platform.openai.com/traces with an OpenAI key,  
or simply remove the `with trace(...)` wrapper — the agent runs identically either way.
